1\. Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2\. Load Dataset

```python
df = pd.read_csv('creditcard.csv')
print(df.head())
print(df['Class'].value_counts())
```

3\. Visualize Class Imbalance

```python
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.show()
```

4\. Amount patterns in fraud vs non fraud

```python
fraud = df[df['Class'] == 1]
non_fraud = df[df['Class'] == 0]

plt.figure(figsize=(10,5))
sns.histplot(non_fraud['Amount'], bins=50, color='green', label='Non-Fraud', kde=True)
sns.histplot(fraud['Amount'], bins=50, color='red', label='Fraud', kde=True)
plt.legend()
plt.title("Transaction Amount Distribution")
plt.show()
```

5\. PCA Plots

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='V1', y='V2', hue='Class', data=df.sample(10000))
plt.title('PCA Component Visualization (V1 vs V2)')
plt.show()
```
